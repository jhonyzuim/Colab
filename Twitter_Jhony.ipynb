```
# Monitorar os tweets
class MyStreamListener(tweepy.StreamListener):
  c = 0
  def on_status(self, status):
    MyStreamListener.c += 1
    # salvar em um arquivo
    # salvar em um bd NoSQL,
    # salvar em planilha eletr√¥nica

    texto = status.text
    print(MyStreamListener.c, status.created_at,texto)

    with open(f'/content/drive/MyDrive/MBA/Web Mining e Social Network Analysis/tweets_{status.id}', 'w') as arq:
      arq.write(texto)

    if (MyStreamListener.c >= 10):
      return False

  def on_error(self, status_code):
    print(f'Erro: {status_code}')
```

```
myStreamListener = MyStreamListener()
myStream = tweepy.Stream(auth=api.auth, listener=myStreamListener)

myStream.filter(track=['bbb'],languages=['pt'])
```

```
# an√°lise de sentimento em tweets
import nltk
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('stopwords')
nltk.download('rslp')
nltk.download('punkt')
```

```
base_tweets = pd.read_csv('/content/drive/MyDrive/MBA/Web Mining e Social Network Analysis/tweets_sentiment/tweets_sentiment.csv', encoding='utf8')
```

```
base_tweets = pd.concat([
    base_tweets[base_tweets.sentiment=='Positivo'].sample(frac=0.01, random_state=88),  
    base_tweets[base_tweets.sentiment=='Negativo'].sample(frac=0.01, random_state=88),                   
])
base_tweets.sentiment.value_counts()
```

```
stemmer = nltk.stem.RSLPStemmer()

def stemmer_tokenizer_sentiment(doc):
  tokens = [t.lower() for t in nltk.tokenize.TweetTokenizer().tokenize(doc)]
  tokens = [stemmer.stem(t) for t in tokens]
  return tokens

stemmer_tokenizer_sentiment('üò≠14üò© 23 intelig√™ncia artificial üòíüòî:)')
```

```
stopwords = [stemmer.stem(s) for s in nltk.corpus.stopwords.words('portuguese')]
```

```
with open('/content/drive/MyDrive/MBA/Web Mining e Social Network Analysis/tweets_sentiment/words_neg.txt', 'r',) as f:
  words_neg = f.read().replace('\n', ' ')

with open('/content/drive/MyDrive/MBA/Web Mining e Social Network Analysis/tweets_sentiment/words_pos.txt', 'r',) as f:
  words_pos = f.read().replace('\n', ' ')

vectorizer_pos_neg = TfidfVectorizer(
    binary=True,
    norm=None,
    use_idf=False,
    stop_words = stopwords,
    tokenizer = stemmer_tokenizer_sentiment
)

vetor_pos_neg = vectorizer_pos_neg.fit_transform([words_neg, words_pos])
vetor_pos_neg
```

```
vetor_tweets = vectorizer_pos_neg.transform(base_tweets['tweet_text'])
```

```
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(vetor_tweets, vetor_pos_neg)
print(sim.shape)
print(sim)
```

```
y_predict = ['Negativo' if s[0] > s[1] else 'Positivo' for s in sim]
y_predict
```

```
from sklearn.metrics import accuracy_score
acuracia = accuracy_score(base_tweets['sentiment'], y_predict)
acuracia
```

```
p = pd.DataFrame(data=y_predict, columns=['sentimento'])
p = pd.DataFrame(p.value_counts())
p
```
