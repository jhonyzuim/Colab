```
import datetime
from datetime import datetime
```

```
df['datetime']=df['order_purchase_timestamp'].apply(lambda row : datetime.strptime(str(row),'%Y-%m-%d %H:%M:%S'))

```

```
df['weekday']=df['datetime'].apply(lambda row: row.weekday())
```

```
df['dayhour']=df['datetime'].apply(lambda row: row.hour)
```
