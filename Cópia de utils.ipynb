```
#@title Coleta das revisões do Kindle separadas por estrelas { vertical-output: true }
import os, time, random
from bs4 import BeautifulSoup
import requests

def salva_arquivo(
    pasta, nome, conteudo, 
    encoding='utf8'):
  # cria pasta, caso não exista
  if not os.path.isdir(pasta):
    os.makedirs(pasta, exist_ok=True)

  # quantidade de arquivos salva, para nomear o próximo
  n_files = len(os.listdir(pasta))+1

  # escrever no arquivo
  with open(
      f'{pasta}/{nome}_{n_files}.txt',
      'w', encoding=encoding) as file:
    file.write(conteudo)

def crawler_kindle(base_url, i, n):
  
  if i<=n:
    print(i)
    response = requests.get(f'{base_url}{i}')
    reviews = BeautifulSoup(response.text, 'html.parser')(attrs={'data-hook':'review'})
    

    for r in reviews:
      # extrair quantidade de estrelas
      stars = r(class_='a-link-normal')[0].text.strip()[0]
      # extrair titulo
      title = r(attrs={'data-hook':'review-title'})[0].text.strip()
      # extrair texto da revisão
      text = r(attrs={'data-hook':'review-body'})[0].text.strip()

      salva_arquivo(f'/content/drive/My Drive/Colab Notebooks/nlp/data/kindle_review/{stars}', 'kindle', f'{title} {text}')

      # se conseguiu coletar, incrementa para a próxima página
    if reviews:
      i+=1
    # espera um tempo para não sobrecarregar a página
    time.sleep(random.randint(1, 10))
    crawler_kindle(base_url, i, n)

```

```
def retorna_textos(pasta):
    lista_conteudos = []
    # lista os arquivos da pasta
    for a, arquivo in enumerate(os.listdir(pasta)):
      #abre o arquivo
      with open(f'{pasta}/{arquivo}', 'r', encoding="utf8") as arq:
          # adiciona na lista para retornar
          lista_conteudos.append(arq.read())
    return lista_conteudos
```
